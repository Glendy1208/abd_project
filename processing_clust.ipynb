```python
import pandas as pd

# Load the provided CSV files
data_egg_fix_path = r"C:\Users\GLENDY\Kuliah\Semester5\analisis_big_data\uas\data\Data_Egg_Fix_cleaned.csv"
clustered_data_path = r"C:\Users\GLENDY\Kuliah\Semester5\analisis_big_data\uas\data\Hasil_Clustered_Per_Tahun_Cleaned.csv"

# Load the datasets
data_egg = pd.read_csv(data_egg_fix_path)
clustered_data = pd.read_csv(clustered_data_path)

# Display the first few rows of each dataset to understand their structure
data_egg.head(), clustered_data.head()
```

```python
# Merge the two datasets on 'Provinsi' to combine production and clustering data
merged_data = pd.merge(data_egg, clustered_data, on="Provinsi", suffixes=("_production", "_cluster"))

# Display the first few rows of the merged dataset to verify the structure
merged_data
```

```python
# Convert the production columns to integers
production_columns = ['2018_production', '2019_production', '2020_production', '2021_production', '2022_production', '2023_production']
merged_data[production_columns] = merged_data[production_columns].astype(int)

# Display the updated dataframe to verify the changes
merged_data
```

```python
from sklearn.impute import KNNImputer

# Inisialisasi KNNImputer
imputer = KNNImputer(missing_values=0, n_neighbors=5)

# Lakukan imputasi pada kolom produksi
merged_data[production_columns] = imputer.fit_transform(merged_data[production_columns])

# Tampilkan dataframe yang telah diimputasi
merged_data
```

```python
# Convert the production columns to integers
production_columns = ['2018_production', '2019_production', '2020_production', '2021_production', '2022_production', '2023_production']
merged_data[production_columns] = merged_data[production_columns].astype(int)

# Display the updated dataframe to verify the changes
merged_data
```

```python
merged_data.to_csv('data/egg_clustering_data.csv', index=False)
```
